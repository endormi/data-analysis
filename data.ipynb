```python
countries_data = pd.read_csv('data/countries.csv')
```

```python
countries_data
```

```python
countries_data[countries_data.country == 'Finland']
```

```python
fi = countries_data[countries_data.country == 'Finland']
```

```python
swe = countries_data[countries_data.country == 'Sweden']
```

```python
swe
```

```python
plt.plot(fi.year, fi.population / 10**6)
plt.plot(swe.year, swe.population / 10**6)
plt.legend(['Finland', 'Sweden'])
plt.xlabel('year')
plt.ylabel('Population')
plt.show()
```

```python
CA = countries_data[countries_data.country == 'Canada']
```

```python
US = countries_data[countries_data.country == 'United States']
```

```python
plt.plot(CA.year, CA.population / 10**6)
plt.plot(US.year, US.population / 10**6)
plt.legend(['United States', 'Canada'])
plt.xlabel('year')
plt.ylabel('Population')
plt.show()
```
