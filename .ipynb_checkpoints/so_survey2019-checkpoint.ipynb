```python
import pandas as pd
```

```python
survey_data = pd.read_csv('data/survey_results2019.csv')
```

```python
survey_data.head(10)
```

```python
survey_data['LanguageWorkedWith'].head(10)
```

```python
languages = survey_data['LanguageWorkedWith'] \
    .str.split(';', expand=True)
languages.head(10)
```

```python
summary = languages.apply(
    pd.Series.value_counts)
summary.head(15)
```

```python
dev_type = survey_data['DevType'] \
    .str.split(';', expand=True)
dev_type.head(10)
```

```python
country = survey_data['Country'] \
    .str.split(';', expand=True)
country.head(10)
```

```python
edu = survey_data['EdLevel'] \
    .str.split(';', expand=True)
edu.head(50)
```

```python
summary = pd.DataFrame({'count': 
    summary.sum(axis=1)
    .groupby(lambda x: x.strip())
    .sum()})
summary
```

```python
total = survey_data[survey_data['LanguageWorkedWith'].notnull()].shape[0]
summary['percent'] = summary['count']/total*100
summary
```

```python
summary.plot.barh(y='percent', figsize=(15,15))
```
