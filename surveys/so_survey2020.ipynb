# Stack overflow 2020 developer survey
## Analysing Developer type, Education level and Most Common Language used by developers

```python
import pandas as pd
survey_data = pd.read_csv('survey_results2020.csv')
survey_data.head(10)
```

```python
summ = pd.DataFrame({'count': 
    summ.sum(axis=1)
    .groupby(lambda x: x.strip())
    .sum()})
summ
```

```python
tot = survey_data[survey_data['EdLevel'].notnull()].shape[0]
summ['percent'] = summ['count']/tot*100
summ
```

```python
summ.plot.barh(y='percent', figsize=(5,5))
```
